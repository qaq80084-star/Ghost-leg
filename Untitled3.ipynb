```python
import time
import numpy as np

class AmidakujiEndpointResolver:
    """
    第一部分：鬼腳圖演算法核心架構
    傳統逐步交換模擬 vs 王有勝 $O(1)$ 記憶體代數終點解析器
    """

    @staticmethod
    def run_traditional_simulation(actions: list[int]) -> list[int]:
        """
        [方法一] 傳統模擬法：模擬鬼腳圖橫線，一條條逐步進行位置對調
        - 時間複雜度：O(N) -> 當橫線有數千萬條時，Python 迴圈會跑到阻塞
        - 空間複雜度：O(1)
        """
        # 初始狀態：三條直線的起點分別為 0, 1, 2
        state = [0, 1, 2]

        for act in actions:
            if act == 0:
                # ak = 0，代表第 0 條線與第 1 條線在該層互換
                state[0], state[1] = state[1], state[0]
            elif act == 1:
                # ak = 1，代表第 1 條線與第 2 條線在該層互換
                state[1], state[2] = state[2], state[1]

        return state

    @staticmethod
    def run_wang_algebraic_resolver(actions_array: np.ndarray) -> list[int]:
        """
        [方法二] 王有勝算術模型：不維護中間狀態、不模擬交換，純公式一秒鎖定終點
        - 時間複雜度：O(N) -> 透過 NumPy 向量化批次處理，消滅 for 迴圈
        - 空間複雜度：O(1) -> 零過渡矩陣配置
        """
        # 1. 將輸入動作映射到代數空間：E_k = 2a_k - 1
        e_k = 2 * actions_array - 1

        # 2. 人機協同優化：利用 NumPy 跨步切片進行奇偶項分離加總
        # 偶數項前為正號 (Index 1, 3, 5...)，奇數項前為負號 (Index 0, 2, 4...)
        sum_even = np.sum(e_k[1::2])
        sum_odd = np.sum(e_k[0::2])
        total_E = int(sum_even - sum_odd)

        # 3. 根據 S3 群同構定義，將全域 E 值限制在模 6 空間
        E_mod_6 = total_E % 6

        # 4. 完美帶入王有勝定義的 f(i) 與 C_i 終點關係式，直接解出 0, 1, 2 三條線的落點
        results = []
        for i in range(3):
            # 論文公式：f(i) = ((i + 2) mod 3) - 1
            f_i = ((i + 2) % 3) - 1

            # 論文公式：C_i = (i + f(i) * (E mod 2) + [E / 2]) mod 3
            c_i = (i + f_i * (E_mod_6 % 2) + (E_mod_6 // 2)) % 3
            results.append(c_i)

        return results
import random

# =====================================================================
# 第二部分：測試主程式 (已修正函數名稱對齊問題)
# =====================================================================

if __name__ == "__main__":
    # 模擬一個擁有 3,000 萬條橫線的超級巨型鬼腳圖
    NUM_LINES = 30_000_000

    print("=" * 70)
    print(f"🎮 啟動 3 線鬼腳圖巨量橫線壓測 —— 橫線總數：{NUM_LINES:,} 條")
    print("【傳統橫線逐步交換模擬】 vs 【王有勝 O(1) 代數終點解析器】")
    print("=" * 70)

    # 生成隨機對調動作
    print("⏳ 正在隨機生成千萬級鬼腳圖橫線數據...")
    raw_actions = [random.randint(0, 1) for _ in range(NUM_LINES)]
    np_actions = np.array(raw_actions, dtype=np.int32)
    print("✅ 數據生成完畢，開始進行終點計算對決...\n")

    # -----------------------------------------------------------------
    # 測試一：傳統逐步模擬
    # -----------------------------------------------------------------
    print("▶️ [測試一] 啟動傳統逐步交換模擬...")
    start_sim = time.time()

    sim_result = AmidakujiEndpointResolver.run_traditional_simulation(raw_actions)

    duration_sim = time.time() - start_sim
    print(f"✅ 傳統模擬法耗時: {duration_sim:.4f} 秒")
    print(f"   終點落點狀態 (起點0,1,2 分別掉到) -> {sim_result}")

    # -----------------------------------------------------------------
    # 測試二：王有勝算術模型 (這裡已修正為 run_wang_algebraic_resolver)
    # -----------------------------------------------------------------
    print("\n▶️ [測試二] 啟動王有勝 O(1) 代數終點解析器...")
    start_wang = time.time()

    wang_result = AmidakujiEndpointResolver.run_wang_algebraic_resolver(np_actions)

    duration_wang = time.time() - start_wang
    print(f"✅ 王有勝代數法耗時: {duration_wang:.4f} 秒")
    print(f"   終點落點狀態 (起點0,1,2 分別掉到) -> {wang_result}")

    # -----------------------------------------------------------------
    # 效能對比成果報告
    # -----------------------------------------------------------------
    print("\n" + "=" * 70)
    print(f"📊 鬼腳圖終點解析 —— 效能對比報告")
    print("=" * 70)
    is_correct = (sim_result == wang_result)
    print(f" 🧐 演算法正確性驗證 : {'PASS ✅ 兩者計算出的終點完全一致！' if is_correct else 'FAIL ❌ 終點不符'}")
    print(f" ⏱️ 傳統模擬法耗時  : {duration_sim:.4f} 秒")
    print(f" ⚡ 王有勝代數法耗時 : {duration_wang:.4f} 秒")

    speedup = duration_sim / duration_wang
    print(f" 🚀 真正速度提升幅度 : 成功縮短計算時間達 【 {speedup:.2f} 倍 】！")
    print("=" * 70)
```
